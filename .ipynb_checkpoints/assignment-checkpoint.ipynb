```python
import random
import math

#Initial Variables, You can change these
gridx = 10
gridy = 2
obstaclepercent = 10

#amount of obstacles
obstacleamount = math.floor((((gridx * gridy) - 2) * (obstaclepercent/100)))

#symbols
symbolstart = 'S'
symbolend = 'G'
symbolopen = '.'
symbolobstacle = 'H'



#randomise start and end
def randomcoord():
    return (random.randint(0, gridx-1), random.randint(0, gridy-1))
#start and end location
start = randomcoord()
end = randomcoord()
#if the start and end location are the same
while start == end:
    end = randomcoord()

#start = (6, 9)
#end = (4, 5)

#start = (5, 3)
#end = (7, 8)

#solution path
solutionpath = list(tuple())

#if the goal is near the solution
flag = False
if start[0] == end[0]:
    if end[1] > 0:
        if start[1] == (end[1] + 1):
            flag = True
    if end[1] < gridy:
        if start[1] == (end[1] - 1):
            flag = True
if start[1] == end[1]: 
    if end[0] > 0:
        if start[0] == (end[0] + 1):
            flag = True
    if end[0] < gridx:
        if start[0] == (end[0] - 1):
            flag = True
            
            
#finds a solution path. simple rightangled approach
if flag == False:
    if start[0] < end[0]:
        for x in range(start[0]+1,end[0]+1):
            solutionpath.append((x,start[1]))
    else:
        for x in range(start[0]-1,end[0]-1,-1):
            solutionpath.append((x,start[1]))
    if abs(start[1] - end[1]) >= 1:
        if start[1] < end[1]: 
            for y in range(start[1],end[1]-1):
                solutionpath.append((end[0],y+1))
        else:
            for y in range(start[1],end[1]+1,-1):
                solutionpath.append((end[0],y-1))
                
    else:
        solutionpath = solutionpath[:-1]
            


    
#defining the grid and adding the start/end
thegrid = [[symbolopen]*gridy for i in range(gridx)]    
thegrid[start[0]][start[1]] = symbolstart
thegrid[end[0]][end[1]] = symbolend




obstaclearray = []
while obstacleamount > 0:
    temp = randomcoord()
    aflag = False
    #check if obstacle already exists
    for coord in obstaclearray:
        if coord == temp:
            aflag = True
    #check if in solution path
    for coord in solutionpath:
        if coord == temp:
            aflag = True
    #check if it equal to the start or end
    if temp == start or temp == end:
        aflag = True
    #if all these check are confirmed than add obstacle
    if aflag == False:
        obstaclearray.append(temp)
        obstacleamount -= 1

        
        
        
#visual for solution path
#for coord in solutionpath:
    #thegrid[coord[0]][coord[1]] = '*'        
        
#print obstcales in grid        
for coord in obstaclearray:
    thegrid[coord[0]][coord[1]] = symbolobstacle
    

    
#prints grid    
for y in range(0,gridx):
    print(thegrid[y])
    
print(start)
print(end)



```
